### 4- CORRELATION AND HEATMAP  

```python
data.corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(data.corr(), annot= True, cbar= True, cmap= 'BuPu')
```

**The heatmap above shows that there is a significant correlation of the data among the columns.**

For example;

_There is positive correlation between:_
1. the density and the residual_sugar.
2. density and fixed_acidity.
3. fixed_acidity and citric_acid.
4. the free_sulfur_dioxide and total_sulfur_dioxide.

_whereas the 3 below show a negative correlation:_
1. alcohol and density 
2. fixed_acidity and pH
3. citric_acid and volatile_acidity
